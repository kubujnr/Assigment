# Computational Physics

Computational physics is a multidisciplinary field that integrates physics, mathematics, computer science, and numerical techniques to develop and implement computer simulations and mathematical models in order to solve physical problems. Its primary objective is to narrow the gap between theoretical physics, which provides mathematical models for physical systems, and experimental physics, which generates measurements of physical phenomena.

<img src='image/computational_physics.webp' width='300' />

Computational physics involves using numerical algorithms and computer simulations to investigate physical systems that are difficult or impossible to study experimentally, such as the behavior of matter under extreme conditions. It has widespread applications in fields such as astrophysics, condensed matter physics, high-energy physics, and materials science.

In recent times, the advancement of high-performance computing has significantly broadened the scope of computational physics, enabling researchers to simulate more intricate physical systems and conduct more detailed analyses of experimental data. As a result, computational physics has become a crucial tool for advancing our knowledge of the natural world.

<font color="red"><h2> Method and Results </h2></font>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eig

class ParticleInABox:
    def __init__(self, L, m, V):
        self.L = L  # length of the box
        self.m = m  # mass of the particle
        self.V = V  # potential energy function
        
class ParticleInABoxSolver:
    def __init__(self, particle, N):
        self.particle = particle
        self.N = N
        self.hbar = 1.055e-34  # Planck's constant divided by 2π
        self.dx = particle.L / N  # step size
        self.x = np.linspace(0, particle.L, N)  # position array
        self.H = None  # Hamiltonian matrix
        self.E = None  # energy levels
        self.psi = None  # wave functions
        
    def set_hamiltonian_matrix(self):
        self.H = np.zeros((self.N, self.N))
        for i in range(1, self.N-1):
            self.H[i, i] = -2 / (self.dx ** 2) + self.particle.V(self.x[i])
            self.H[i, i-1] = 1 / (self.dx ** 2)
            self.H[i, i+1] = 1 / (self.dx ** 2)
        self.H[0, 0] = -2 / (self.dx ** 2) + self.particle.V(self.x[0])
        self.H[0, 1] = 1 / (self.dx ** 2)
        self.H[-1, -1] = -2 / (self.dx ** 2) + self.particle.V(self.x[-1])
        self.H[-1, -2] = 1 / (self.dx ** 2)
        
    def solve(self):
        self.set_hamiltonian_matrix()
        self.E, self.psi = eig(self.H)
        idx = self.E.argsort()
        self.E = self.E[idx]
        self.psi = self.psi[:, idx]
        
    def plot_wave_functions(self, num_functions=5):
        fig, ax = plt.subplots()
        for i in range(num_functions):
            ax.plot(self.x, self.psi[:, i], label=f"n={i+1}")
        ax.set_xlabel("x")
        ax.set_ylabel("ψ(x)")
        ax.legend()
        plt.show()
        
    def plot_energy_levels(self):
        fig, ax = plt.subplots()
        ax.plot(range(1, len(self.E)+1), self.E, 'bo')
        ax.set_xlabel("n")
        ax.set_ylabel("Energy (J)")
        ax.set_title("Particle in a one-dimensional box")
        plt.show()

V = lambda x: np.inf if x < 0 or x > 1 else 0

particle = ParticleInABox(1, 1, V)
solver = ParticleInABoxSolver(particle, N=100)
solver.solve()
solver.plot_wave_functions(num_functions=5)
solver.plot_energy_levels()
```
